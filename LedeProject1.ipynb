### **Step 8:** Create a code to separate states/cities/regionsfrom their country so they don't appear as countries themselves.

```python
#country / international chart code for datawrapper. Need to separate US states from USA as country for different charts. 

#when I ran my code and put it into datawrapper it showed certain locations such as LA, Greater
#^Cincinatti, Western Australia, etc. as countries rather than locations in a country which is why I went back and altered it manually/created these lists. 

us_states = [ #all caps like in HTML.
    "CALIFORNIA", "FLORIDA", "TEXAS", "NEW YORK", "ARIZONA",
    "COLORADO", "WASHINGTON", "NEVADA", "ILLINOIS", "UTAH",
    "MISSOURI", "MICHIGAN", "LOUISIANA", "OREGON", "MINNESOTA",
    "TENNESSEE", "GEORGIA", "OHIO", "PUERTO RICO","DC", "D.C."
]

region_map = {
    "WESTERN AUSTRALIA": "AUSTRALIA",
    "ONTARIO": "CANADA",
    "GAUTENG": "SOUTH AFRICA",
    "GREATER CINCINNATI": "USA",
    "LOS ANGELES": "USA"
}

countries = []

for loc in df["location"]:
    loc_upper = str(loc).upper()

    # 1. USA detection for states
    is_us = any(state in loc_upper for state in us_states)

    if is_us:
        countries.append("USA")

    else:
        # 2. If format is "City, Country/Region"
        if ", " in loc_upper:
            last_part = loc_upper.split(", ")[-1].strip()

            # fix known provinces/regions
            if last_part in region_map:
                countries.append(region_map[last_part])
            else:
                countries.append(last_part)

        # 3. 
        else:
            countries.append(loc_upper)

df["country"] = countries
```

### **Step 9:** Create a column for openings per year 

```python
#create a year column:

df["year"] = df["date"].str.extract(r"(\d{4})") #used regex  

#create chart data for datawrapper - show openings per year from 2003 - 2026: 

year_df = df.groupby("year").size().reset_index(name="openings") #reset_index makes new column/df & .size() counts rows in groups.

year_df #displays table w/ above work done.
```

### **Step 12:** Sort US states to see the number of grand openings in each

```python
#need to separate US states from the country list and have info for state chart since most concentrated in USA.

us_rows = []

for loc in df["location"]:
    loc_upper = str(loc).upper()

    for state in us_states:
        if state in loc_upper:
            us_rows.append(state)

state_df = pd.Series(us_rows).value_counts().reset_index()
state_df.columns = ["state", "count"]

state_df
```

### **Step 13:** Group the openings for each country per year into 3 time periods for datawrapper chart

```python
#below groups it into the time periods I assign:
#need to make year an int not str
df["year"] = df["year"].astype(int)

def get_period(year):
    if 2003 <= year <= 2009:
        return "2003–2009"
    elif 2010 <= year <= 2016:
        return "2010–2016"
    elif 2017 <= year <= 2026:
        return "2017–2026"
    else:
        return None

#below makes new column before making pivot table for time periods above 
df["period"] = df["year"].apply(get_period)
```

```python
#google / TA said to do pivot table 
table = df.pivot_table(
    index="country",
    columns="period",
    values="title",
    aggfunc="count",
    fill_value=0
)

table
```
